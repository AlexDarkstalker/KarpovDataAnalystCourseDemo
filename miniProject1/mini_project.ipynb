```python
hotel_df = pd.read_csv('https://stepik.org/media/attachments/lesson/360344/bookings.csv', encoding='windows-1251', sep=";")
```

```python
hotel_df.columns = hotel_df.columns.str.lower().str.replace(' ', '_')
```

```python
hotel_df[hotel_df['is_canceled'] == 0] \
                                    .groupby('country') \
                                    .aggregate({'is_canceled': 'count'}) \
                                    .sort_values(by='is_canceled', ascending=False)
```

```python
hotel_df.hotel.unique()
```

```python
hotel_df.groupby('hotel', as_index=False) \
        .aggregate({'stays_total_nights':'mean'}) \
        .style.format({'stays_total_nights': '{:.2f}'.format})
```

```python
hotel_df.columns
```

```python
hotel_df[hotel_df['reserved_room_type'] != hotel_df['assigned_room_type']].shape
```

```python
hotel_df[hotel_df['arrival_date_year'] == 2016] \
                    .groupby('arrival_date_month') \
                    .aggregate({'hotel': 'count'}) \
                    .sort_values(by='hotel', ascending=False)
```

```python
hotel_df[hotel_df['arrival_date_year'] == 2017] \
                    .groupby('arrival_date_month') \
                    .aggregate({'hotel': 'count'}) \
                    .sort_values(by='hotel', ascending=False)
```

```python
hotel_df.groupby('arrival_date_year')['arrival_date_month'].value_counts()
```

```python
hotel_df[['adults', 'children', 'babies']].mean()
```

```python
hotel_df['total_kids'] = hotel_df.children + hotel_df.babies
```

```python
hotel_df.groupby('hotel') \
        .aggregate({'total_kids': 'mean'}) \
        .style.format({'total_kids': '{:.2f}'.format})
```

```python
hotel_df['has_kids'] = hotel_df['total_kids'] > 0
```

```python
(hotel_df.groupby('has_kids')['is_canceled'].value_counts(normalize=True) * 100).apply('{:.2f}'.format)
```
